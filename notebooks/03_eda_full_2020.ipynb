```python
# üì¶ Imports
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# üìÑ Chargement du fichier optimis√©
df = pd.read_parquet("../data/raw/optimized_2020.parquet")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Taux de valeurs manquantes
(df.isnull().mean() * 100).sort_values(ascending=False).head(20)
```

## Analyse des valeurs manquantes
1. Colonnes tr√®s peu renseign√©es (> 90%)

| Colonne                                                     | Taux de NaN | Action recommand√©e                                             |
| ----------------------------------------------------------- | ----------- | -------------------------------------------------------------- |
| `ancien_code_commune`, `lotX_surface_carrez`, `lotX_numero` | > 90 %      | **√Ä supprimer** |

2. Colonnes partiellement manquantes (10% √† 60%)

| Colonne                             | Taux de NaN | Action                                                                    |
| ----------------------------------- | ----------- | ------------------------------------------------------------------------- |
| `surface_reelle_bati`                                       | 59 %        | Critique pour calcul prix/m¬≤ ‚Üí **filtrer les lignes nulles**   |
| `nombre_pieces_principales`, `code_type_local`              | 43 %        | M√™me traitement : **supprimer ou √©tudier si rempla√ßables**     |
| `adresse_numero`, `surface_terrain` | 30‚Äì40 %     | Peu critiques ‚Üí √† **garder si utiles**, sinon ignorer                     |


3. Colonnes faiblement incompl√®tes (< 5%)

| Colonne                                              | Taux de NaN | Action                                                       |
| ---------------------------------------------------- | ----------- | ------------------------------------------------------------ |
| `valeur_fonciere`, `code_postal`, `adresse_nom_voie` | < 2%        | Tr√®s exploitables ‚Üí **conserver**, filtrer les NaN √† l‚Äôusage |
| `latitude`, `longitude`             | 3.1 %       | G√©olocalisation tr√®s utile ‚Üí **garder**, filtrer les nulls ponctuellement |

```python
### visualisation des r√©partitions des donn√©es num√©riques
# S√©lection des colonnes num√©riques (hors ID ou g√©o)
num_cols = df.select_dtypes(include=['int32', 'float32']).columns.tolist()
num_cols = [col for col in num_cols if col not in ['longitude', 'latitude']]  # √† exclure si besoin
num_cols
```

```python
df['valeur_fonciere_log'] = np.log1p(df['valeur_fonciere'])
sns.histplot(df['valeur_fonciere_log'], bins=50)
plt.title("Log(valeur fonci√®re + 1)")
plt.show()
```

```python
# Affichage group√© des histogrammes
n = len(num_cols)
n_cols = 3
n_rows = (n // n_cols) + 1

plt.figure(figsize=(n_cols * 5, n_rows * 4))

for i, col in enumerate(num_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.histplot(df[col].dropna(), bins=20, kde=False)
    plt.title(col)
    plt.tight_layout()
```

```python
# Liste des colonnes √† supprimer (remplissage < 10%)
cols_to_drop_1 = [
    'ancien_code_commune', 'ancien_nom_commune', 'ancien_id_parcelle',
    'lot1_numero', 'lot1_surface_carrez', 'lot2_numero', 'lot2_surface_carrez',
    'lot3_numero', 'lot3_surface_carrez', 'lot4_numero', 'lot4_surface_carrez',
    'lot5_numero', 'lot5_surface_carrez'
]
# Suppression des colonnes
df.drop(columns=cols_to_drop_1, inplace=True)

# V√©rification des colonnes restantes
print(f"Shape after dropping columns: {df.shape}")
```

```python
# Liste des colonnes avec infos inutiles
cols_to_drop_2 = [
    'numero_disposition', 'adresse_numero', 'adresse_suffixe',
    'adresse_nom_voie', 'adresse_code_voie', 'nom_commune','nature_culture_speciale'
    'numero_volume', 'code_nature_culture', 'nature_culture','code_nature_culture_speciale'
]
df = df.drop(columns=cols_to_drop_2, axis=1)
print(f"Shape after dropping more columns: {df.shape}")
```

## Etude des variables cat√©gorielles


```python
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
cat_cols = [col for col in cat_cols if col not in ['id_mutation', 'id_parcelle']] 
print(f"{len(cat_cols)} colonnes cat√©gorielles :")
print(cat_cols)
```

```python
# R√©partition
for col in cat_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False).head(10))
```

```python
# Nombre max de modalit√©s √† afficher
max_modalities = 30

for col in cat_cols:
    nb_modalities = df[col].nunique()
    if nb_modalities <= max_modalities:
        plt.figure(figsize=(10, 4))
        sns.countplot(data=df, x=col, order=df[col].value_counts().index)
        plt.title(f"R√©partition des modalit√©s ‚Äì {col}")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```
